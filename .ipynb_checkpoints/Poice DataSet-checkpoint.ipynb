## Police DataSet


    Here,
    
    The data from police Check Post is given.
    
    This data is available as CSV file. We are going to analyze this data set using Pandas DataFrame.

```python
import pandas as pd
```

```python
data = pd.read_csv("Police Data.csv")
```

```python
data.head()
```

## Instruction (For Data Cleaning)

1. Remove the coloumn that only contains missing value.

```python
data.isnull().sum()
```

```python
data.drop( columns = 'country_name', inplace=True)
```

```python
data.head()
```

## Question(Base on filtering + Value Count)

2. For Speeding, were Men or Women stoped more often?

```python
data.head()
```

```python
data[data.violation == 'Speeding'].driver_gender.value_counts()
```

## Question (Groupby)

3. Does gender affect who get searched during a stop ? 

```python
data.head()
```

```python
data.groupby('driver_gender').search_conducted.sum()
```

# Question (mapping + data-type casting)

4. What is means the stop_duration ? 

```python
data.head()
```

```python
data.stop_duration.value_counts()
```

```python
data['stop_duration'] = data['stop_duration'].map({'0-15 Min':7.5, '16-30 Min':24, '30+ Min':45})
```

```python
data.head()
```

```python
data['stop_duration'].mean()
```
